# Books Recommender System using clustering | collaborative based

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
users = pd.read_csv('data/BX-Users.csv', sep= ";" ,on_bad_lines='skip' , encoding = 'latin-1')
```

```python
books = pd.read_csv('data/BX-Books.csv', sep= ";" ,on_bad_lines='skip' , encoding = 'latin-1')
```

```python
books.head(2)
```

```python
books = books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.head(3)
```

```python
users.shape
```

```python
users.head()
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv', sep= ";" ,on_bad_lines='skip' , encoding = 'latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={
    "User-ID": "user_id",
    "Book-Rating": "rating"
}, inplace=True)
```

```python
ratings
```

```python
x = ratings["user_id"].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings["user_id"].isin(y)]
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books, on ="ISBN" )
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.head(2)
```

```python
ratings_with_books.rename(columns={
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": "publisher"
},inplace=True)
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={
    "rating": "num_of_ratings"
}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on = 'title')
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_ratings']>=50]
final_rating.head()
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(["user_id","title"], inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values ='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
pip install scikit-learn
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance , suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance , suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    for i in range(len(suggestion)):
        book = book_pivot.index[suggestion[i]]
        for j in book:
            print(j)
```

```python
book_name = "A Bend in the Road"
recommend_book(book_name)
```
